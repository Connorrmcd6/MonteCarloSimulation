```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
def Pi_approx_plot(n):
    plt.rcParams["figure.figsize"] = (5,5)
    circle_count=0
    xvals=[]
    yvals=[]
    for i in range(0, n+1):
        x= random.uniform(-1,1)
        y= random.uniform(-1,1)
        xvals.append(x)
        yvals.append(y)
        
    for j in range(0, n+1):
        D=((xvals[j])**2+(yvals[j])**2)**0.5
        if D<=1:
            plt.plot(xvals[j],yvals[j],'ro')
            circle_count= circle_count+1
            
        else:
            plt.plot(xvals[j],yvals[j],'bo')
            
    plt.axis('square')
    plt.show()
    act= np.pi
    app=4*circle_count/n
    a= abs(act-app)/act
    print('Number of Points=', n)
    print('Pi Approximation=', app) 
    print('Actual Pi value=', act)
    print('% Error =', a*100)
    
```

```python
Pi_approx_plot(5000)
```

```python
def Pi_no_plot(n):
    circle_count=0
    xvals=[]
    yvals=[]
    for i in range(0, n+1):
        x= random.uniform(-1,1)
        y= random.uniform(-1,1)
        xvals.append(x)
        yvals.append(y)
        
    for j in range(0, n+1):
        D=((xvals[j])**2+(yvals[j])**2)**0.5
        if D<=1:
            circle_count= circle_count+1
            
    act= np.pi
    app=4*circle_count/n
    a= abs(act-app)
    return app

def Pi_line(n):
    plt.rcParams["figure.figsize"] = (15,10)
    plx=[1,n]
    ply=[np.pi, np.pi]
    vals=[]
    x=[]
    for i in range(1,n+1):
        v=Pi_no_plot(i)
        vals.append(v)
        x.append(i)
    
    #return vals
    plt.plot(x, vals)
    plt.plot(plx, ply, 'r-')
    print('Approximate Value of Pi after {} steps='.format(n), vals[n-1])
```

```python
Pi_line(5000)
```
